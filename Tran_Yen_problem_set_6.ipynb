###1

```python
import pandas as pd
from sklearn.datasets import load_diabetes
diabetes = load_diabetes()
X = pd.DataFrame(diabetes.data,columns=diabetes.feature_names)
y = diabetes.target
```

###2

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

###3

```python
from sklearn.model_selection import train_test_split
#a
def model(X,y, check):
  #b
  Xtrain, Xtest, ytrain, ytest = train_test_split(X,y,random_state=1)
  model = LinearRegression(fit_intercept=True)
  #c
  model.fit(Xtrain,ytrain)
  #d
  y_test_model = model.predict(Xtest)
  y_train_model = model.predict(Xtrain)

  print('Training R^2',model.score(Xtrain,ytrain))
  print('Testing R^2',model.score(Xtest,ytest))
  print('y = ',model.coef_[0],'*x + ',model.intercept_)

  if check == True:
    graph_model(Xtrain, Xtest, ytrain, ytest, y_train_model,y_test_model)
```

###4

```python
model(X[['bp']],X[['s6']],False)
```

```python
model(X[['age']],X[['s6']],False)
```

```python
model(X[['bmi']],X[['s6']],False)
```

###5

```python
def graph_model(Xtrain, Xtest, ytrain, ytest, y_train_model, y_test_model,):

    fig, (ax1,ax2) = plt.subplots(1,2,figsize=[20,10])
    if Xtrain.shape[1] == 1:
      ax1.scatter(Xtrain[Xtrain.columns[0]],ytrain)
      ax1.plot(Xtrain[Xtrain.columns[0]],y_train_model)
     
    if Xtest.shape[1] == 1:
      ax2.scatter(Xtest[Xtest.columns[0]],ytest)
```

###6

```python
for x in X.columns:
  model(X[[x]],X[['s6']],True)
```
