###Importing libaries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

###Webscraping data

```python
webdata1 = pd.read_html('https://en.wikipedia.org/wiki/MonsterVerse')
webdata2 = pd.read_html('https://en.wikipedia.org/wiki/List_of_Fast_%26_Furious_characters')
monster_df = webdata1[3]
fast_df = webdata2[2]
```

###Defining function and clean tables for MonsterVerse

```python
def clean_table(x,is_monster):

  if is_monster == True:
    x.columns = x.columns.droplevel(0)
    x.columns = x.columns.droplevel(1)

  if is_monster == False:
    x.columns = x.columns.droplevel(0)
    x.columns = x.columns.droplevel(2)
    x.columns = x.columns.droplevel(1)

  x.index = x['Character']
  x.drop(columns='Character',inplace=True)

  if is_monster == True:
    x = x.drop(index='Humans')
  
  else:
    x = x.drop(index = 'Primary supporting characters')

  return x
```

```python
monster_df = clean_table(monster_df,True)
```
