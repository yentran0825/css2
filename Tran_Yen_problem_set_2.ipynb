###Block 1: Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = sns.load_dataset('anscombe')
```

###Block 2: Creating the visualization

```python
I = data[data['dataset'] == 'I']
II = data[data['dataset'] == 'II']
III = data[data['dataset'] == 'III']
IV = data[data['dataset'] == 'IV']
ds_list = [I ,II, III, IV]
```

```python
data
```

```python
fig = plt.figure()
fig,axes = plt.subplots(nrows=2,ncols=2, sharex=True, sharey=True, figsize=(10,10))
axes = axes.flatten()

title = ['I','II','III','IV']
style = ['r', 'g', 'b', 'm']
count = 0;


for a in axes:
  a.plot(ds_list[count]['x'], ds_list[count]['y'], linestyle='', marker = 'o', color = style[count])
  a.set_xlim([0,20])
  a.set_ylim([0,15])
  a.set_xlabel('x')
  a.set_ylabel('y')
  a.set_title(title[count])
  x = ds_list[count]['x']
  y = ds_list[count]['y']
  stats = (f'$\\mu_x$ = {np.mean(x):.2f}\n'
            f'$\\mu_y$ = {np.mean(y):.2f}\n'
             f'$\\sigma_x$ = {np.std(x):.2f}\n'
              f'$\\sigma_y$ = {np.std(y):.2f}\n')

  bbox = dict(boxstyle='round', fc='blanchedalmond', ec='orange', alpha=0.5)
  a.text(0.95, 0.07, stats, fontsize=11, bbox=bbox,transform=a.transAxes, horizontalalignment='right')

  count+=1
```

###Block 3: Saving the image

```python
fig.savefig('fig1.png')
```
